```python
import pandas as pd

#import dataframe
df = pd.read_csv('DW3_set_exercise.csv')

print(df.head())
```

```python
#Define Code Lists
diabetes = ['E08','E09','E10','E11','E13']
covid = ['U07.1','J12.82']
```

```python
#Diabetes Patients / Cardinality
diabetes_df = df[df['Diagnosis Code'].isin(diabetes)]
diabetes_unique_patients = diabetes_df['Patient ID'].unique().tolist()
print(f"Diabetes patient IDs:", diabetes_unique_patients)
print(f"Cardinality of Diabetes dataset:", len(diabetes_unique_patients))
```

```python
#COVID Patients / Cardinality
covid_df = df[df['Diagnosis Code'].isin(covid)]
covid_unique_patients = covid_df['Patient ID'].unique().tolist()
print(f"Covid patient IDs:", covid_unique_patients)
print(f"Cardinality of Covid dataset:", len(covid_unique_patients))
```

```python
intersect_df = pd.merge(diabetes_df, covid_df, how='inner', left_on='Patient ID', right_on='Patient ID')
intersect = intersect_df['Patient ID'].unique().tolist()
print(f"Patients with BOTH covid and diabetes:", intersect)
print("Cardinality of intersection set:", len(intersect))
```

```python
unioned_patient_ids = covid_unique_patients + diabetes_unique_patients
unioned_count = len(diabetes_unique_patients) + len(covid_unique_patients)
print(f"Patients with covid OR diabetes:", set(unioned_patient_ids))
print("Cardinality of intersection set:", len(set(unioned_patient_ids)))
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn3, venn3_circles

set_covid = set(covid_unique_patients)
set_diab = set(diabetes_unique_patients)
set_both = set(intersect)
set_combined = set(unioned_patient_ids)

intersect_v = set_covid & set_diab
only_c = set_covid - set_diab
only_d = set_diab - set_covid
only_comb = set_combined - set_both

venn3(subsets=(len(only_c), len(only_d), len(intersect_v), 
               len(only_comb), len(set_both), 0, 0),
      set_labels=('COVID', 'Diabetes', 'Both'))

venn3_circles(subsets=(len(only_c), len(only_d), len(intersect_v), 
               len(only_comb), len(set_both), 0, 0))
plt.show()
```

```python
import numpy as np
df_after = df[df['Patient ID'].isin(intersect)]
df_after['flag'] = np.where(df_after['Diagnosis Code'].isin(covid), 'c',
                             np.where(df_after['Diagnosis Code'].isin(diabetes), 'd', 'o'))
grouped_codes = df_after.groupby('Patient ID')['Diagnosis Code'].agg(list).reset_index()
df_after2 = df_after[['Patient ID', 'Date', 'flag']]
pivot_df_after = df_after2.pivot_table(index='Patient ID', columns='flag', values='Date', aggfunc='first').reset_index()
pivot_df_after = pd.merge(pivot_df_after, grouped_codes, on='Patient ID')
d_after_c = pivot_df_after[pivot_df_after['d'] > pivot_df_after['c']]
d_after_c_patients = d_after_c['Patient ID'].tolist()
codes = [i for list in d_after_c['Diagnosis Code'] for i in list]
counts = pd.Series(codes).value_counts()
filtered_counts = counts[diabetes]

print(f"Patients with diabetes AFTER covid:", d_after_c_patients)
print("Cardinality of diabetes AFTER covid set:", len(d_after_c_patients))
print(filtered_counts)
```
